```python
#cantidad de datos faltantes de cada columna
dat_vacios = expvida.isnull().sum()
print(dat_vacios)
```

```python
#maximo valor de datos vacios
maxdatvacios = expvida.isnull().sum().max()
print(f"El valor máximo de datos vacios es {maxdatvacios}")
```

```python
#cantidad de filas total y divido el maximo valor de datos vacios por este numero y div por 100
filas_total = len(expvida)
resultado = (maxdatvacios / filas_total) * 100
#muestra el número completo
print(resultado)  
#muestra 2 decimales
print(f"El porcentaje máximo de filas nulas es igual a {resultado:.2f} %")  
#guarda el número formateado como string
por_str = f"{resultado:.2f}"  
#imprime "22.19"
print(por_str)  
#imprime <class 'str'>
print(type(por_str))  
```

```python
#renombro columnas del dataframe
expvida = expvida.rename(columns={"Life expectancy ": "life_expectancy", "Measles": "measles"})
print(expvida.columns)
```

```python
#promedio de la columna
expvida["life_expectancy"].mean()
```

```python
#promedio por pais
expvida.groupby('Country')['life_expectancy'].mean()
```

```python
#maxima expectativa de vida usando pandas
maxima_exp = expvida['life_expectancy'].max()
print(maxima_exp)
```

```python
#maxima expectativa de vida usando numpy
maxima_exp = np.nanmax(expvida['life_expectancy'].values)
print(maxima_exp)
```

```python
#minima expectativa de vida usando pandas
minima_exp = expvida['life_expectancy'].min()
print(minima_exp)
```

```python
#minima expectativa de vida usando numpy
minima_exp = np.nanmin(expvida['life_expectancy'].values)
print(minima_exp)
```

```python
#muestro resultados maximos
fila_maxima = expvida.loc[expvida['Country'].idxmax()]
print(f"La fila con el valor máximo es: {fila_maxima}")
```

```python
#traigo valores de la columna 'status'
valores_unicos = expvida['Status'].unique()
print(valores_unicos)
```

```python
#cuantos paises hay en la columna 'status'
expvida.groupby('Status')['Country'].value_counts()
```

```python
#cuantas filas tienen valores distintos
cantidad_filas = (expvida['Hepatitis B'] != 0).sum()
print(cantidad_filas)
```
